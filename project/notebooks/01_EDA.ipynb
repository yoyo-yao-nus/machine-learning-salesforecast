```python
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
# plt.figure(figsize=(6,6))
# plot the scatter matrix among highly correlated columns
highcor_columns = ['clickVolume', 'ma14SalesVolume','sales']
scatter_matrix(df[highcor_columns],figsize=(10,10))
```

```python
columns = ['avgOriginalUnitPrice','avgFinalUnitPrice']
plt.figure(figsize=(6,6))
plt.scatter(list(df[columns[0]]),list(df[columns[1]]))
plt.xlabel(columns[0])
plt.ylabel(columns[1])
plt.savefig('../figures/avg2.png')
```
